<strong>Выписать несколько членов последовательности $x^{(n)}$, $n = 1, 2, 3, \dots$. Установить, является ли последовательность $x^{(n)}$ фундаментальной и сходящейся в указанном пространстве. <font color='red'>(3 балла за каждый пункт)</font></strong>

a) $l^1$, $x^{(n)} = (\frac{2}{n^4}, \frac{4}{n^4}, \frac{6}{n^4}, \dots, \frac{2n}{n^4}, 0, 0, \dots)$

Пусть $x^{(n)} = (x_1^{(n)}, x_2^{(n)}, x_3^{(n)}, \dots, x_k^{(n)}, \dots)$.

По условию
$x_k^{(n)} =
\begin{cases} 
\frac{2k}{n^4}, & \mbox{if 1 $\leqslant$ } k \mbox{ $\leqslant$ n}
\\ 0, & \mbox{if } k\mbox{ > n} 
\end{cases}$

Выпишем несколько членов последовательности $x^{(n)}$ при $n = 1, 2, 3, 4$:
$$x^{(1)} = (2, 0, 0, \dots)$$
$$x^{(2)} = (\frac{2}{16},\frac{4}{16}, 0, 0, \dots)$$
$$x^{(3)} = (\frac{2}{81}, \frac{4}{81}, \frac{6}{81}, 0, 0, \dots)$$
$$x^{(4)} = (\frac{2}{256}, \frac{4}{256}, \frac{6}{256}, \frac{8}{256}, 0, 0, \dots)$$

Очевидно, что для каждого фиксированного $k$ имеем $\frac{2k}{n^4} \to 0$ при $n \to \infty$. Поэтому можно предположить, что пределом $\{x^{(n)}\}_{n=1}^{\infty}$ является нулевая последовательность в пространстве $l^1$.

<u>Гипотеза</u>: $x^{(n)} \to 0$ в пространстве $l^1$ при $n \to \infty$.

Проверим гипотезу по определению сходимости в пространстве $l^1$:

$\rho_{l^1}(x^{(n)}, 0) = \sum_{k=1}^{\infty} |{x_k^{(n)} - 0}| = \frac{2 + 4 + 6 + \dots + 2n}{n^4} = 2\frac{1 + 2 + 3 + \dots + n}{n^4} = \frac{2(1 + n)n}{2n^4} = \frac{(1 + n)}{n^3} \to 0$ при $n \to \infty$.

Приходим к выводу, что $\{x^{(n)}\}_{n=1}^{\infty}$ - сходящаяся последовательность $x^{(n)} \to 0$ при $n \to \infty$ в пространстве $l^1$. Сходящаяся последовательность является также и фундаментальной. 

b) $l^\infty$, $x^{(n)} = (n, n - 1, n - 2, \dots, 2, 1, 0, 0, \dots)$

Пусть $x^{(n)} = (x_1^{(n)}, x_2^{(n)}, x_3^{(n)}, \dots, x_k^{(n)}, \dots)$.

По условию 
$x_k^{(n)} =
\begin{cases} 
n - (k - 1), & \mbox{if 1 $\leqslant$ } k \mbox{ $\leqslant$ n}
\\ 0, & \mbox{if } k\mbox{ > n} 
\end{cases}$

Выпишем несколько членов последовательности $x^{(n)}$ при $n = 1, 2, 3, 4$:
$$x^{(1)} = (1, 0, 0, \dots)$$
$$x^{(2)} = (2, 1, 0, 0, \dots)$$
$$x^{(3)} = (3, 2, 1, 0, 0, \dots)$$
$$x^{(4)} = (4, 3, 2, 1, 0, 0, \dots)$$

Отсюда видно, что каждая из координат $x_k^{(n)}$ неограниченно возрастает при $n \to \infty$

Действительно, для каждого фиксированного $k$ имеем $x_k^{(n)} = n - (k - 1) \to \infty$ при $n \to \infty$.

<u>Гипотеза</u>: $x^{(n)}$ не имеет предела в пространстве $l^\infty$.
Чтобы окончательно доказать, что последовательность $x^{(n)}$ не обладает сходимостью в пространстве $l^\infty$, покажем, что она даже не является фундаментальной в этом пространстве:

$\rho_{l^\infty}(x^{(n)}, x^{(m)}) = \sup_{1 \leqslant k < \infty}{|x^{(n)} - x^{(m)}|} = 
|(n - k - 1) - (m - k - 1)| = \sup_{1 \leqslant k < \infty}{|n - m|} = |n - m|$

Нельзя утверждать, что $|n - m| \to 0$ при $n, m \to \infty$. Действительно, выражение $|n - m|$ представляет собой неопределенность вида $[\infty - \infty]$ и может иметь разный предел при разном соотношении между числами $n$ и $m$. Например, если взять $n = 2m$, то $|n - m| = |m| \to \infty$ при $m, n \to \infty$

Таким образом, последовательность $x^{(n)}$ не фундаментальная в пространстве $l^\infty$ и тем более не является сходящейся в этом пространстве.
